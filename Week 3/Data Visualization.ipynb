```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("iris.csv")  # Change path if needed

# Optional: Drop 'Id' column if not needed
df.drop(columns=['Id'], inplace=True)
print(df.head())
```

```python
# 1. Histogram of all numeric features
df.hist(figsize=(10, 6), edgecolor='black')
plt.suptitle('Histograms of Iris Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter Plot: Sepal Length vs Petal Length with Species hue
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Petal Length', y='Sepal Length',)
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```
