```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv")

df.rename(columns={
    "Sepal Length (cm)": "SepalLength",
    "Sepal Width (cm)": "SepalWidth",
    "Petal Length (cm)": "PetalLength",
    "Petal Width (cm)": "PetalWidth"
}, inplace=True)

# Drop ID column
df.drop("Id", axis=1, inplace=True)

df.head()
```

```python
sns.pairplot(df, hue="Species", diag_kind="kde")
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()

# Shows pairwise relationships and class separability visually.
```

```python
#Scatter Plot: Sepal Length vs Petal Length
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PetalLength', y='SepalLength', hue='Species')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(data=df, x="Species", y="PetalLength", jitter=True, size=5)
plt.title("Strip Plot of Petal Length by Species")
plt.ylabel("Petal Length (cm)")
plt.show()
#Shows how individual data points vary within each species class.
```

```python
plt.figure(figsize=(8, 5))
sns.swarmplot(data=df, x="Species", y="PetalWidth")
plt.title("Swarm Plot of Petal Width by Species")
plt.ylabel("Petal Width (cm)")
plt.show()
# Preserves point separation; useful for small datasets like Iris.
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Species", palette="pastel")
plt.title("Count of Each Iris Species")
plt.show()
#Confirms class balance (all 3 species have 50 samples).
```

```python
g = sns.FacetGrid(df, col="Species", height=4)
g.map_dataframe(sns.scatterplot, x="PetalLength", y="SepalWidth")
g.set_axis_labels("Petal Length", "Sepal Width")
g.fig.suptitle("Petal Length vs Sepal Width by Species", y=1.05)
plt.show()
# compare patterns across species in separate subplots.
```

```python
plt.figure(figsize=(6, 5))
sns.kdeplot(data=df, x="PetalLength", y="PetalWidth", hue="Species", fill=True, thresh=0.05)
plt.title("2D KDE Plot: Petal Length vs Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
# Smooth heatmap-style density, reveals natural clustering per species.
```

```python

features = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]

# Compute the correlation matrix
correlation_matrix = df[features].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```
