**Titanic Dataset - Exploratory Data Analysis (EDA)**

1. Load the Titanic Dataset

```python
import pandas as pd
df = pd.read_csv('tested.csv')
df.head()
```

2. Check Data Summary and Missing Values

```python
df.info()
df.describe(include='all')
```

 Missing Value Detection


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap="Blues")
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap="Blues")
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)
```

3. Univariate Analysis

Categorical: Sex, Embarked, Pclass

```python
sns.countplot(x='Sex', data=df)
sns.countplot(x='Pclass', data=df)
```

Numerical: Age, Fare

```python
sns.histplot(df['Age'], bins=30, kde=True)
sns.histplot(df['Fare'], bins=40, kde=True)
```

4. Bivariate Analysis


Survival vs Gender/Class

```python
sns.barplot(x='Sex', y='Survived', data=df)
sns.barplot(x='Pclass', y='Survived', data=df)
```

Boxplots for distribution comparison

```python
sns.boxplot(x='Survived', y='Age', data=df)
```

5. Insights from Analysis
✅ Findings
Females survived more.

First-class passengers had better chances.

Age and fare showed interesting trends.



6. Survival Analysis

```python
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack')
sns.histplot(data=df, x='Fare', hue='Survived', multiple='stack')
```

7. Feature Engineering

Create FamilySize Feature

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

IsAlone Feature

```python
df['IsAlone'] = 1 # default
df.loc[df['FamilySize'] > 1, 'IsAlone'] = 0
```

8.Advanced Visualization (Heatmap)

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```
